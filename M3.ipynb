```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
print(tips)
```

```python
#sns.barplot(data=tips,x='day',y='total_bill',ci=None)
sns.barplot(data=tips,x='day',y='total_bill',hue="time")
plt.title("Average bill amount per day")
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
plt.title("Average bill amount per day")
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='smoker')
plt.title("bill by day - smoker VS non-smoker")
plt.show()
```

```python
sns.violinplot(data=tips,x='day',y='total_bill')
plt.title("Bill distribution shape by day")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.title("Tip VS Total bill")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="time")
plt.title("Tip VS Bill - Lunch VS Dinner")
plt.show()
```

```python
sns.regplot(data=tips,x="total_bill",y="tip")
plt.title("Tip VS Bill + best fit line")
plt.show()
```

```python
corr =  tips.select_dtypes(include="number").corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation between numbers")
plt.show()
```

```python
sns.histplot(data=tips,x="total_bill")
plt.title("How bills are distributed")
plt.show()
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",col="time",hue="smoker",style="smoker",size="size")
```

```python
sns.lineplot(data=tips,x="day",y="total_bill",hue="time",style="sex")
plt.title("Average bill amount by day of week")
plt.ylabel("Average total bill")

```
