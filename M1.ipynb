```python
# Prime number
prime = []
num = 2

while len(prime) < 10:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        prime.append(num)
    num += 1

print("First 10 prime numbers:", prime)
```

```python
# Fibonacci series
n = 10  
a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# Factorial
def factorial(n):
    
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
num = 3
print("Factorial of", num, "is", factorial(num))
```
